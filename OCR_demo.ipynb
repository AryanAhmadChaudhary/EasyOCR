```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
''''Single Line Text Detection'''
```

```python
Image_PATH = "surf.jpeg"
```

```python
reader = easyocr.Reader(['en'], gpu = False)
result = reader.readtext(Image_PATH)
result
```

```python
top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(Image_PATH)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left,font,0.5,(0,0,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
''''Multiple Line Text Detection'''
```

```python
Image_PATH = "sign.png"
```

```python
reader = easyocr.Reader(['en'], gpu = False)
result = reader.readtext(Image_PATH)
result
```

```python
img = cv2.imread(Image_PATH)
spacer = 100
for detection in result:
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
    img = cv2.putText(img,text,(20,spacer), font, 0.5,(0,255,0),2,cv2.LINE_AA)
    spacer+=15

plt.imshow(img)
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
Image_PATH = "//content/ocr_sample.jpg"
reader = easyocr.Reader(['en'], gpu = False)
result = reader.readtext(Image_PATH)
# result
# Read the image
img = cv2.imread(Image_PATH)
spacer = 100
extracted_texts = []  # List to store all extracted texts

# Iterate through the detections
for detection in result:
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]

    # Draw a rectangle around the detected area
    # img = cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 3)
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)

    # Add the detected text to the image
    # img = cv2.putText(img, text, (20, spacer), font, 0.5, (255, 255, 0), 2, cv2.LINE_AA)
    img = cv2.putText(img,text,top_left,font,0.5,(0,0,255),2,cv2.LINE_AA)
    spacer += 15

    # Append the detected text to the list
    extracted_texts.append(text)

# Display the image with detections
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# Print or use the extracted texts array
# print(extracted_texts)
for text in extracted_texts:
    print(text)
```
